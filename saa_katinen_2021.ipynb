# Sää päivittäin

Tuodaan tarvittavat kirjastot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Luetaan tiedot

```python
data = pd.read_csv('saa_katinen_pv2021.csv')
data.head(12)
```

Tiedot muuttujiin

```python
vuosi = data['Vuosi']
kuukaudet = data['Kk']
paiva = data['Pv']
df = pd.DataFrame({'day': paiva,'month': kuukaudet,'year': vuosi})
df["date"] = pd.to_datetime(df)
lampotilat = data['Ilman lämpötila (degC)']
print(df)
```

Rakennetaan kaavio

```python
plt.figure(figsize=(20,5))
plt.plot(df["date"],lampotilat)
plt.ylabel('Lämpötila (°C)')
plt.title('Lämpötila Katisissa 2021')
plt.grid(axis = 'y', linestyle = '--')
plt.show()
```
