```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('saa_katinen_pv2021.csv')
data.head(12)
```

```python
vuosi = data['Vuosi']
kuukaudet = data['Kk']
paiva = data['Pv']
df = pd.DataFrame({'day': paiva,'month': kuukaudet,'year': vuosi})
df["date"] = pd.to_datetime(df)
print(df)
lampotilat = data['Ilman lämpötila (degC)']
```

```python
plt.figure(figsize=(20,5))
plt.plot(df["date"],lampotilat)
#plt.xlabel('Kuukausi')
plt.ylabel('Lämpötila (°C)')
plt.title('Lämpötila Katisissa 2021')
#labels = ['Tammikuu','Helmikuu','Maaliskuu','Huhtikuu','Toukokuu','Kesäkuu','Heinäkuu','Elokuu','Syyskuu','Lokakuu','Marraskuu','Joulukuu']
#plt.xticks(range(1,365,7))
plt.grid(axis = 'y', linestyle = '--')
plt.show()
```
